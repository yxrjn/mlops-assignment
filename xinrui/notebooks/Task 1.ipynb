```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load the Wheat Seeds dataset
wheat_seeds_data = pd.read_csv("03_Wheat_Seeds.csv")
```

```python
## Wheat Seeds Dataset
# Check the first few rows of the dataset
wheat_seeds_data.head()
```

```python
# Display summary statistics
print("Summary statistics:")
print(wheat_seeds_data.describe())
```

```python
wheat_seeds_data.head()
```

```python
# Wheat Seeds Dataset
# Check data types and missing values
wheat_seeds_data.info()

# Check for missing values
wheat_seeds_data.isnull().sum()
```

```python
# Correlation matrix heatmap to examine relationships between numerical features
plt.figure(figsize=(10, 8))
sns.heatmap(wheat_seeds_data.corr(), annot=True, cmap='coolwarm', linewidths=0.5)
plt.title('Correlation Heatmap for Wheat Seeds Dataset')
plt.show()
```

```python
# Pairplot to visualize relationships between features and how they are grouped by 'Type'
sns.pairplot(wheat_seeds_data, hue='Type', palette='Set1')
plt.suptitle('Pairplot for Wheat Seeds Dataset (Grouped by Type)', y=1.02)
plt.show()
```

```python
# Boxplot for each feature grouped by Type (to see distributions of features for each variety of wheat)
features = ['Area', 'Perimeter', 'Compactness', 'Length', 'Width', 'AsymmetryCoeff', 'Groove']
for feature in features:
    plt.figure(figsize=(8,6))
    sns.boxplot(x='Type', y=feature, data=wheat_seeds_data)
    plt.title(f'{feature} vs Wheat Type')
    plt.xlabel('Type of Wheat')
    plt.ylabel(feature)
    plt.show()
```

```python
# Violin plots for each feature grouped by Type to visualize density
for feature in features:
    plt.figure(figsize=(8,6))
    sns.violinplot(x='Type', y=feature, data=wheat_seeds_data, inner='quart')
    plt.title(f'{feature} vs Wheat Type (Violin Plot)')
    plt.xlabel('Type of Wheat')
    plt.ylabel(feature)
    plt.show()
```

```python
# Histograms for each numerical feature to inspect distributions
wheat_seeds_data[features].hist(bins=15, figsize=(15, 10), layout=(2,4))
plt.suptitle('Histograms for Wheat Seeds Features')
plt.show()
```

```python
# Distribution of the target variable 'Type'
plt.figure(figsize=(8,6))
sns.countplot(x='Type', data=wheat_seeds_data)
plt.title('Distribution of Wheat Types')
plt.xlabel('Type of Wheat')
plt.ylabel('Count')
plt.show()
```

```python
# Scatter plot: Area vs Perimeter grouped by Type
plt.figure(figsize=(8,6))
sns.scatterplot(x='Area', y='Perimeter', hue='Type', palette='Set1', data=wheat_seeds_data)
plt.title('Area vs Perimeter grouped by Type')
plt.xlabel('Area')
plt.ylabel('Perimeter')
plt.show()
```
